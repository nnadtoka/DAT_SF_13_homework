### With Bank.csv dataset

```python
# import pandas
import pandas as pd

# read in bank data
bank = pd.read_csv('bank.csv', sep=';')

# explore data
bank.head()
```

```python
bank_grouped_by_y = bank.groupby('y')
subscribed = bank_grouped_by_y.get_group('yes')
not_subscribed = bank_grouped_by_y.get_group('no')
```

```python
#out of the subscribed what are the counts of poutcome value
subscribed.groupby('poutcome').size()
```

```python
#out of the not subscribed what are the counts of poutcome value
not_subscribed.groupby('poutcome').size()
```

```python
# verify unique values for marital status - a good candidate for dummy variables with only 3 possible values
pd.unique(bank['marital'].ravel())
```

```python
# Create dummy variables for marital status
marital_dummies = pd.get_dummies(bank.marital, prefix='marital').ix[:, 1:]

# concatenate the two dummy variable columns onto the original DataFrame
# note: axis=0 means rows, axis=1 means columns
bank = pd.concat([bank, marital_dummies], axis=1)

# print the updated DataFrame
bank.head(10)
```

```python
# unique values for job
pd.unique(bank['job'].ravel())
```

```python
# the list of job titles is too long for dummy variables
# lets map the jobs roughly based on typical income groups
bank['job'] = bank.job.map({'unknown':0,'unemployed':1, 'student':2, 'housemaid':3,
                            'retired':4, 'services':5, 'technician':6, 'admin.':7, 
                            'self-employed':8, 'blue-collar':9, 'entrepreneur':10, 'management':11})
```

```python
# dataset description: https://archive.ics.uci.edu/ml/datasets/Bank+Marketing
# all 4521 values are given for all numeric columns, but we have pdays (-1)
len(bank[bank['pdays']==-1].index)
#3705 values of pdays are set to -1
```

```python
# 0 value of pdays is not currently used
len(bank[bank['pdays']==0].index)
```

```python
# I believe that -1 (instead of 999 mentioned on the website) 
# means client was not previously contacted)
len(bank[bank['pdays']==999].index)
```

```python
# We set pdays that were -1 to 0 (negative values for number of days don't make sense)
bank.pdays[bank['pdays']==-1]=0
# alternative day doing the same thing...
#bank['pdays']=bank['pdays'].replace(to_replace=[-1], value=0)
```

```python
pd.unique(bank.education.ravel())
```

```python
# education has some notion of order, we replace the 
# uknown with 0 and the rest with corresponding numbers 1,2,3
bank['education'] = bank.education.map({'unknown':0, 'primary':1,'secondary':2, 'tertiary':3})
```

```python
pd.unique(bank.month.ravel())
```

```python
# map months to numbers from jan to dec from 0 to 11
bank['month'] = bank.month.map({'jan':0, 'feb':1, 'mar':2, 'apr':3, 'may':4, 'jun':5, 'jul':6, 'aug':7,
                        'sep':8, 'oct':9,'nov':10, 'dec':11})
```

```python
pd.unique(bank['poutcome'].ravel())
# Here we only care about failure and success, lets assign everything else to other
```

```python
# we map other and uknown values of poutcome to 0 and failure to 1 and success to 2
bank['poutcome'] = bank.poutcome.map({'other':0,'unknown':0,'failure':1, 'success':2})
```

```python
# in this case we only have yes,no values for bank loan, replace it with 1 and 0 correspondingly
pd.unique(bank['loan'].ravel())
```

```python
bank['loan'] = bank.loan.map({'no':0,'yes':1})
```

```python
# in this case we only have yes,no values for housing, replace it with 1 and 0 correspondingly
pd.unique(bank['housing'].ravel())
```

```python
bank['housing'] = bank.housing.map({'no':0,'yes':1})
```

```python
# default has only yes and no answers, which we will replace with 1 and 0
pd.unique(bank['default'].ravel())
bank['default'] = bank.default.map({'no':0,'yes':1})
```

```python
pd.unique(bank['contact'].ravel())
```

```python
# contact way does not have any order, lets create some dummy variables

# Create dummy variables for marital status
contact_dummies = pd.get_dummies(bank.contact, prefix='contact').ix[:, 1:]

# concatenate the two dummy variable columns onto the original DataFrame
# note: axis=0 means rows, axis=1 means columns
bank = pd.concat([bank, contact_dummies], axis=1)
```

```python
bank.describe()
```

```python
bank.head()
```

```python
# the target variable only has yes and no, we map it to 1 and 0
bank['y'] = bank.y.map({'no':0, 'yes':1})
```

```python
bank['y'].sum()
```

```python
bank.describe()
```

```python
# create a list of feature columns ( skip duration, 
# since it is not know apriori before we dial the prospect)
# thus not relevant for our prediction
feature_cols = ['age', 'job', 'education', 'balance', 'housing', 'loan', 'day', 
                'month', 'campaign', 'pdays','previous', 'poutcome', 
                'marital_married', 'marital_single']

# define X and y
X = bank[feature_cols]
y = bank.y

# fit a classification tree with max_depth=3 on all data
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=10, random_state=1)
treeclf.fit(X, y)

# compute the feature importances
pd.DataFrame({'feature':feature_cols, 'importance':treeclf.feature_importances_}).sort('importance',ascending=False)
```

```python
feature_cols_crop = ['poutcome', 'balance', 'age', 'month', 'day', 'pdays', 'job', 'previous',
                     'campaign', 'marital_married','education', 'housing']
features = bank[feature_cols_crop].values
target = bank.y.values
```

```python
# create a Graphviz file
from sklearn.tree import export_graphviz
with open("bank.dot", 'wb') as f:
    f = export_graphviz(treeclf, out_file=f, feature_names=feature_cols)

# at the command line, run this to convert to PNG:
# dot -Tpng bank.dot -o bank.png
```

```python
from sklearn.tree import DecisionTreeClassifier
#treeclf = DecisionTreeClassifier(max_depth=3, random_state=0)

treeclf = DecisionTreeClassifier(compute_importances=None, criterion='gini',max_depth=5,
                                 max_features=None,max_leaf_nodes=None,
                                 min_density=None,min_samples_leaf=1,
                                 min_samples_split=2,
                                 random_state=0, splitter='best')

# use cross-validation to find best max_depth
from sklearn.cross_validation import cross_val_score
treeclf.fit(features, target)
scores = cross_val_score(treeclf,features,target, cv =5, scoring = 'accuracy')
scores
```

```python
# create a Graphviz file
from sklearn.tree import export_graphviz
with open("bank.dot", 'wb') as f:
    f = export_graphviz(treeclf, out_file=f, feature_names=feature_cols)

# at the command line, run this to convert to PNG:
# dot -Tpng bank.dot -o bank.png
```

```python
# Conclusion: The prediction accuracy went down dramatically 
# from 0.88 (data.csv) to 0.57 for the larger dataset (bank-additional-full.csv). 
#Possible explanations: 
#1. Maybe the handling of 'unknown', 'no', 'yes' values as numerical 0, 1, 2
# was a bad idea?
# 2. the fact that we introduced the environmental social-economical variables has broken the logic of the model?
```

```python
# Lets assume that introduced social-economical variables break the logic of the model and skip 'euribor3m'

feature_cols_crop =['age','job','campaign','education','day_of_week','housing','poutcome','loan','marital_married',
 'marital_single','default']
features = bank_full[feature_cols_crop].values
target = bank_full.y.values
# max_depth 3 gives better result
from sklearn.tree import DecisionTreeClassifier
#treeclf = DecisionTreeClassifier(max_depth=3, random_state=0)

treeclf = DecisionTreeClassifier(compute_importances=None, criterion='gini',max_depth=3,
                                 max_features=None,max_leaf_nodes=None,
                                 min_density=None,min_samples_leaf=1,
                                 min_samples_split=2,
                                 random_state=0, splitter='best')

# use cross-validation to find best max_depth
from sklearn.cross_validation import cross_val_score

treeclf.fit(features, target)
scores = cross_val_score(treeclf,features,target, cv =5, scoring = 'accuracy')
scores
```
