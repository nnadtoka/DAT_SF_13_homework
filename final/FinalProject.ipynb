*** Target ***

- Identify users actions based on input sensor data

*** Dataset ***

In this dataset, between 77 and 84 sensor data collection boards equipped with reed switch sensors collected data about human activity for two weeks. The sensors were installed in everyday objects such as drawers, refrigerators, containers, etc. to record opening-closing events (activation deactivation events) as the subject carried out everyday activities.
http://courses.media.mit.edu/2004fall/mas622j/04.projects/home/

*** Challenges ***

- Working with sensor data which is not perfectly mapped/annotated to actions
- No control of data quality
- Mulptiple sensors can be activated per single action
- The same sensor can be activated multiple times per single action
- The duration of sensor activation can go past the action duration

```python
# helper function that inserts sensor names and their activation times

def insertIntoDataStruct(aDict, name,startTime,endTime):
    if not name in aDict:
        aDict[name] = [(startTime,endTime)]
    else:
        aDict[name].append((startTime,endTime))
```

### Time manipulation

```python
# helper function to get the period of day
# accepts time object as an input and outputs day-time cathegory
# encoded in integer

def getTimeCat(ts):
    
    hour = ts.hour
    # --> Morning = 0400-1000
    mornStart = 4
    mornEnd = 11

    # --> Midday = 1000-1600
    midStart = 11
    midEnd = 16

    # --> Evening = 1600-2200
    eveStart = 16
    eveEnd = 22

    # --> Late Night = 2200-0400
    lateStart = 22
    lateEnd = 4

    if hour >= mornStart and hour < mornEnd:
      timecat = 0 #morning
    elif hour >= midStart and hour < midEnd:
      timecat = 1 #midday
    elif hour >= eveStart and hour < eveEnd:
      timecat = 2 #evening
    elif hour >= lateStart and hour < lateEnd:
      timecat = 3 #late night
    else:
        timecat = 3

    return timecat
```

### Feature Extraction

```python
from __future__ import division

# Helper function that does heavy lifting to compute features
# The function transforms dictionaries of lists of sensor names and activation times to features
# That are then returned in form of dictionary, easy to convert to python dataframe
def featurizeSensorData(aDict): 
    resDict = {}
    # get activity info
    activity_name = aDict['activity_name']
    activity_start_time = aDict['start_time']
    activity_end_time = aDict['end_time']
    activity_date = aDict['date']
    
    activity_start_date_time = dt.datetime.strptime(activity_date + ' ' + \
                                                      activity_start_time, "%m/%d/%Y %H:%M:%S")
    activity_end_date_time = dt.datetime.strptime(activity_date + ' ' + \
                                                      activity_end_time, "%m/%d/%Y %H:%M:%S")
    
    seconds_activity_duration = (activity_end_date_time - activity_start_date_time).seconds
    
    # remove these keys from the dictionary
    # what we keep are only sensor data
    aDict.pop('activity_name', None)
    aDict.pop('start_time', None)
    aDict.pop('end_time', None)
    aDict.pop('date', None)
    
    resDict['activity_name'] = activity_name
    resDict['activity_start_time'] = activity_start_time
    resDict['activity_end_time'] = activity_end_time
    resDict['activity_date'] = activity_date
    resDict['activity_dayofweek'] = activity_start_date_time.weekday()
    resDict['activity_category'] = getTimeCat(activity_start_date_time)
    # save activity duration
    #l_duration_activity.append(seconds_activity_duration)
    
    for sensor, sensor_activation_times in aDict.iteritems():
        l_delta = []
        l_sec_from_start = []
        l_order = []
        inx = 0
        for start_time, end_time  in sensor_activation_times:
                   
            end_date_time = dt.datetime.strptime(activity_date + ' ' + end_time, \
                                                 "%m/%d/%Y %H:%M:%S")
            start_date_time = dt.datetime.strptime(activity_date + ' ' + start_time, \
                                                   "%m/%d/%Y %H:%M:%S")
            
            # If sensor activation lasts longer than the end of activity, crop it
            # We saw some corner cases and have treatment for them
            if (activity_end_date_time >= end_date_time):
                sensor_act_seconds = (end_date_time - \
                                  start_date_time).seconds
            else:
                if activity_end_date_time > start_date_time:
                    end_date_time = activity_end_date_time
                    sensor_act_seconds = (activity_end_date_time - \
                                  start_date_time).seconds
                else:
                    sensor_act_seconds = (end_date_time - \
                                  start_date_time).seconds
            l_delta.append(sensor_act_seconds)
            
            # get the number of seconds from the start of activity until the sensor activation
            seconds_from_start_activity = (start_date_time - \
                                           activity_start_date_time).seconds
      
            # we had some cases when human annotated activity began 
            # after the sensors associated with activity were activated
            # In this case we set the difference of both (time from activity 
            # start to sensor activation) to 0.0
            if not(seconds_activity_duration > seconds_from_start_activity):
                seconds_from_start_activity = 0.0
            # eventually we may want to set 
            # the activity time will to begin precisely at activation of 
            #the first sensor
            l_sec_from_start.append(seconds_from_start_activity) 
            # we add one to 0 position to distinguish between the given values 
            # and the ones that are not given
            l_order.append(inx+1)
            inx += 1
        #print l_sec_from_start
        #print l_delta
        resDict[str(sensor) + '_active_flag'] = 1
        resDict[str(sensor) + '_active_num'] = len(l_delta)
        
        if len(l_delta)>1:
            resDict[str(sensor) + '_std'] = np.std(l_delta)
        else:   
            resDict[str(sensor) + '_std'] = 0.0
            
        resDict[str(sensor) + '_delta'] = np.mean(l_delta)
        resDict[str(sensor) + '_sec_from_start'] = np.mean(l_sec_from_start)
        resDict[str(sensor) + '_relative_to_start_time'] = np.mean(l_sec_from_start)/seconds_activity_duration
        resDict[str(sensor) + '_relative_order'] = np.mean(l_order)/len(l_delta)
        resDict[str(sensor) + '_order'] = np.mean(l_order)
    
    return resDict
```

### Read the data from the file into dataframe

```python
# in this paragraph we read features from the file
# following the certain known structure of the input file
# we transform file input into features

d = collections.defaultdict(list) 

activities = []
target = []
activity_date = []
activity_start_time = []
activity_end_time = []
sensors = []
sensor_names = []
activation_start_time = []
activation_end_time = []
sensorsCounts = []

d = collections.defaultdict(list) 
mydict = {}
activity_dict = {}

from collections import Counter


mydict_test = {}

inx = 0
activity_inx = 0
with open('activities_data.csv') as file:
    while inx < (int(number_lines) - 1):
        line1 = file.readline()
        line2 = file.readline()
        line3 = file.readline()
        line4 = file.readline()
        line5 = file.readline()
        
        inx += 5
        if not (line1 and line2 and line3 and line4 and line5):
            print "Failed to get a line from the input, breaking..."
            print inx
            break
        else:
            # remove some carriage returns and newlines, split by comma
            activity_info = line1.strip('\r\n').split(",")
            activity_name = activity_info[0] # Bathing
            activity_date = activity_info[1] # 4/1/2003
            activity_start_time = activity_info[2] # 20:41:35
            activity_end_time =  activity_info[3] # 21:32:50
       
            # remove some carriage returns and newlines, split by comma
            # 100,68,81,101,93,137,93,58,57,67,93,58,68,88,57,67,100,68,67,76
            sensors_numbers = line2.strip('\r\n').split(",") 

            #Toilet Flush,Sink faucet - hot,Closet,Light switch, 
            #Shower faucet,Freezer,Shower faucet,Medicine cabinet,
            #Medicine cabinet,Cabinet,Shower faucet,Medicine cabinet,
            #Sink faucet - hot,Sink faucet - cold,Medicine cabinet,Cabinet,
            #Toilet Flush,Sink faucet - hot,Cabinet,Lamp
            
            # remove some carriage returns and newlines, split by comma
            sensors_name_list = line3.strip('\r\n').split(",")         
            activation_start_time = line4.strip('\r\n').split(",")    
            activation_end_time = line5.strip('\r\n').split(",")    
            
            sensors = {}
            sensors['activity_name'] = activity_name
            sensors['date'] = activity_date
            sensors['start_time'] = activity_start_time
            sensors['end_time'] = activity_end_time
            
            for n in range(len(sensors_numbers)):
                insertIntoDataStruct(sensors, sensors_numbers[n], activation_start_time[n],\
                                 activation_end_time[n])
                
            resdict = featurizeSensorData(sensors)
            mydict[activity_inx] = resdict
            activity_inx += 1
            
my_df = pd.DataFrame(mydict).transpose()
```

### KNN

```python
from sklearn import neighbors, datasets, feature_selection
from sklearn.cross_validation import train_test_split, cross_val_score

# Various variables we'll need to set intially.
n_neighbors = range(1, 51, 2)
print n_neighbors

X_train, X_test, y_train, y_test = train_test_split(features,target, test_size=0.3, random_state=12)

# Loop through each neighbors value from 1 to 51 and append
# the scores
scores = []
for n in n_neighbors:
    clf = neighbors.KNeighborsClassifier(n)
    clf.fit(X_train, y_train)
    scores.append(clf.score(X_test, y_test))
```
