*** Target ***

- Identify users actions based on input sensor data

*** Dataset ***

In this dataset, between 77 and 84 sensor data collection boards equipped with reed switch sensors collected data about human activity for two weeks. The sensors were installed in everyday objects such as drawers, refrigerators, containers, etc. to record opening-closing events (activation deactivation events) as the subject carried out everyday activities.
http://courses.media.mit.edu/2004fall/mas622j/04.projects/home/

*** Challenges ***

- Working with sensor data which is not perfectly mapped/annotated to actions
- No control of data quality
- Mulptiple sensors can be activated per single action
- The same sensor can be activated multiple times per single action
- The duration of sensor activation can go past the action duration

```python
import collections

activities = []
target = []
activity_date = []
activity_start_time = []
activity_end_time = []
sensors = []
sensor_names = []
activation_start_time = []
activation_end_time = []
sensorsCounts = []

d = collections.defaultdict(list) 
mydict = {}
mydict_test = {}
from collections import Counter

inx = 0
with open('activities_data.csv') as file:
    while 1:
        line = file.readline()
        if not line:
            break
        elif line and line[0].isalpha():
            line = line.strip('\r\n')
            line_parts = line.split(",")
            # we are at the Activity description example:
            # Bathing,4/1/2003,20:41:35,21:32:50
            if len(line_parts)>=2 and (not(line_parts[0][0].isdigit()) and (line_parts[1][0].isdigit())):
                my_activity_name = line_parts[0] # Bathing
                my_activity_date = line_parts[1] # 4/1/2003
                my_activity_start_time = line_parts[2] # 20:41:35
                my_activity_end_time = line_parts[3]  # 21:32:50
                
                activities.append(my_activity_name)
                target.append(my_activity_name) # Bathing
                activity_date.append(my_activity_date) 
                activity_start_time.append(line_parts[2]) 
                activity_end_time.append(line_parts[3]) # 21:32:50
                lineSensors = file.readline() # 100,68,81,101,93,137,93,58,57,67,93,58,68,88,57,67,100,68,67,76
                if not lineSensors:
                    break
                else:
                    listdict = []
                    lineSensors = lineSensors.strip('\r\n') # remove some carriage returns and newlines
                    listSensors = lineSensors.split(",") # split by comma
                    d[inx].append(listSensors)
                    sensors.append(listSensors)
                    c = Counter(listSensors)
                    sensorsCounts.append(c.items())
                    current_dict = dict(c.items())
                   
                    mydict[(my_activity_date,my_activity_name, my_activity_start_time, my_activity_end_time)] = current_dict
                    current_dict['target'] = my_activity_name
                    current_dict['activity_date'] = my_activity_date
                    current_dict['activity_start_time'] = my_activity_start_time
                    current_dict['activity_end_time'] = my_activity_end_time
                    
                    mydict_test[inx] = current_dict  
                    
                    #get another line
                    lineSensorNames = file.readline() #Toilet Flush,Sink faucet - hot,Closet,Light switch,Shower faucet,Freezer,Shower faucet,Medicine cabinet,Medicine cabinet,Cabinet,Shower faucet,Medicine cabinet,Sink faucet - hot,Sink faucet - cold,Medicine cabinet,Cabinet,Toilet Flush,Sink faucet - hot,Cabinet,Lamp

                    if not lineSensorNames:
                        break
                    else:
                        lineSensorNames = lineSensorNames.strip('\r\n') # remove some carriage returns and newlines
                        listSensorNames = lineSensorNames.split(",") # split by comma
                        sensor_names.append(listSensorNames)
                        #get another line with start of activity times
                        #20:51:52,20:51:58,20:53:36,20:53:49,20:53:52,20:58:22,20:58:43,21:5:23,21:5:46,21:5:47,21:18:34,21:18:55,21:19:41,21:20:4,21:20:38,21:20:39,21:21:13,21:21:16,21:21:37,21:22:8

                        lineStartTime = file.readline() 
                        if not lineStartTime:
                            break
                        else:
                            lineStartTime = lineStartTime.strip('\r\n') # remove some carriage returns and newlines
                            listStartTime = lineStartTime.split(",") # split by comma
                            activation_start_time.append(listStartTime)
                         #get another line with end of activity times
                        lineEndTime = file.readline() 
                        if not lineEndTime:
                            break
                        else:
                            lineEndTime = lineEndTime.strip('\r\n') # remove some carriage returns and newlines
                            listEndTime = lineEndTime.split(",") # split by comma
                            activation_end_time.append(listEndTime) 
                    inx+=1
    
```

```python
# need to find the durations
import datetime as dt
#activation_start_time
activation_start_date_time = []
activation_end_date_time = []
activation_delta = []
activation_seconds_from_start = []
activation_order = []
activity_duration = []
#activation_end_time
#sensor_names
#activity_date
#activity_start_time
#activity_end_time

#sensor_names

# We had some unusual cases in the data, where the sensor activation time (e.g. lamp in the bathroom)
# was greater (a couple of extra hours) than duration of actual activity
# there are multiple ways to treat this case
# Here we chose to crop the end time of activation sensor to the time of the end of activity
# there were some corner cases in the data, hense some special treatment

for ix in range(len(sensor_names)):
    b = activation_start_time[ix]
    c = activation_end_time[ix]
    a = activity_date[ix]
    l_start = []
    l_end = []
    l_delta = []
    l_sec_from_start = []
    l_duration_activity = []
    for n in range(len(b)):
        
        sensor_act_start_date_time = dt.datetime.strptime(a + ' ' + b[n], "%m/%d/%Y %H:%M:%S")
        sensor_act_end_date_time = dt.datetime.strptime(a + ' ' + c[n], "%m/%d/%Y %H:%M:%S")
        activity_start_date_time = dt.datetime.strptime(a + ' ' + \
                                                      activity_start_time[ix], "%m/%d/%Y %H:%M:%S")
        activity_end_date_time = dt.datetime.strptime(a + ' ' + \
                                                      activity_end_time[ix], "%m/%d/%Y %H:%M:%S")
        # If sensor activation lasts longer than the end of activity, crop it
        # We saw some corner cases and have treatment for them
        if (activity_end_date_time >= sensor_act_end_date_time):
            sensor_act_seconds = (sensor_act_end_date_time - \
                                  sensor_act_start_date_time).seconds
        else:
            if activity_end_date_time > sensor_act_start_date_time:
                sensor_act_end_date_time = activity_end_date_time
                sensor_act_seconds = (activity_end_date_time - \
                                  sensor_act_start_date_time).seconds
            else:
                sensor_act_seconds = (sensor_act_end_date_time - \
                                  sensor_act_start_date_time).seconds
              
        l_start.append(sensor_act_start_date_time)
        l_end.append(sensor_act_end_date_time)
        l_delta.append(sensor_act_seconds)
        
        #print sensor_act_seconds, sensor_act_start_date_time, \
        #sensor_act_end_date_time, sensor_names[ix][n],\
        #activities[ix], activity_start_time[ix], activity_end_time[ix]

        # get the number of seconds from the start of activity until the sensor activation
        #if (sensor_act_start_date_time >= activity_start_date_time):
        seconds_from_start_activity = (sensor_act_start_date_time - \
                                           activity_start_date_time).seconds
        seconds_activity_duration = (activity_end_date_time - activity_start_date_time).seconds
        # save activity duration
        l_duration_activity.append(seconds_activity_duration)
        
        # we had some cases when human annotated activity began 
        # after the sensors associated with activity were activated
        # In this case we set the difference of both (time from activity start to sensor activation) to 0.0
        if not(seconds_activity_duration > seconds_from_start_activity):
            seconds_from_start_activity = 0.0
            # eventually we may want to set 
            # the activity time will to begin precisely at activation of the first sensor
        l_sec_from_start.append(seconds_from_start_activity)    
        
    activation_seconds_from_start.append(l_sec_from_start)    
    activation_start_date_time.append(l_start)
    activation_end_date_time.append(l_end)
    activation_delta.append(l_delta)
    activity_duration.append(l_duration_activity)
    # order of the sensor activation in the sequence
    activation_order.append(range(len(b)))
```
